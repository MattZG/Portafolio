# Modelización no supervisada

Durante la etapa de EDA se menciono la posibilidad de utilizar algoritmos no supervisados para este problema y al igual que con los supervisados se busca ver la compasión entre los metodos de optimización. Inicialmente, tenia la intención de aplicar KMeans al dataset pero dado que el numero de valores unicos de la variable target es elevado (5000 valores app) es probable que el modelo no funcione como se espera.

Dicho lo anterior se va a utilizar el algoritmo de *GaussianMixture*. Al igual que KMeans asigna los resultados a un unico cluster, la diferencia radica en que este algoritmo lo hace mediante la probabilidad del punto en pertecener a cada cluster.

## 1. Importar los paquetes 

```python
import pandas as pd
import time
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import adjusted_rand_score, silhouette_score
import matplotlib.pyplot as plt
```

## 2. Carga de los datos

```python
ruta = 'C:/Users/matia/OneDrive/Escritorio/Mass_Prediction'
```

### 2.1 Datos de entrenamiento

```python
datos = 'trabajo.csv'
ruta_completa = ruta + '/02_Datos/03_Trabajo/' + datos
df_train = pd.read_csv(ruta_completa, index_col = 0, delimiter = ',')
df_train.head()
```

### 2.2 Datos de validación

```python
validacion = 'validacion.csv'
ruta_completa = ruta + '/02_Datos/02_Validacion/' + validacion
df_val = pd.read_csv(ruta_completa, index_col = 0, delimiter= ',')
df_val.head()
```

## 3. Preprocesamiento

### 3.1 Limpiar espacios variables

```python
df_train.columns = df_train.columns.str.strip()
df_val.columns = df_val.columns.str.strip()
```

### 3.2 Selección de variables

```python
cols = ['E1', 'E2', 'px1', 'py1', 'pz1', 'px2', 'py2', 'pz2', 
            'pt1', 'pt2', 'eta1', 'eta2', 'phi1', 'phi2', 'Q1', 'Q2', 'M']
```

### 3.3 Duplicados

```python
df_train = df_train.drop_duplicates()
df_val = df_val.drop_duplicates()
```

### 3.4 Pipeline del preprocesamiento

```python
preprocesamiento_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Imputar nulos con la media
    ('scaler', StandardScaler())  # Normalizar los datos
])
```

Aplicacion de la pipeline.

```python
df_train[cols] = preprocesamiento_pipeline.fit_transform(df_train[cols])
df_val[cols] = preprocesamiento_pipeline.transform(df_val[cols])
```

## 4. Separación de variables

```python
train_x = df_train.drop(columns = 'M')
train_y = df_train['M']
val_x = df_val.drop(columns = 'M')
val_y = df_val['M']
```

## 5. Modelización

Instanciar

```python
model = GaussianMixture(random_state = 42)
```

Definir los parametros

```python
grid = {
    'n_components': [2, 3, 4],
    'covariance_type': ['full', 'tied', 'diag', 'spherical'],
    'tol': [1e-3, 1e-4, 1e-5],
    'reg_covar': [1e-3, 1e-2, 1e-1]
}
```

Entrenar el modelo con *GridSearchCV*

```python
start_time = time.time()
grid_search = GridSearchCV(model, grid, cv = 3, n_jobs = -1, error_score = 'raise')
grid_search.fit(train_x)
grid_time = time.time() - start_time
```

```python
best_model_grid = grid_search.best_estimator_ 
best_model_grid
```

```python
start_time = time.time()
random_search = RandomizedSearchCV(model, param_distributions = grid, n_iter = 3, cv = 5, n_jobs = -1, random_state = 42)
random_search.fit(train_x)
random_time = time.time() - start_time
```

```python
best_model_random = random_search.best_estimator_
best_model_random
```

```python
y_pred_grid = best_model_grid.predict(val_x)
y_pred_random = best_model_random.predict(val_x)
```

```python
silhouette_grid = silhouette_score(val_x, y_pred_grid)
silhouette_random = silhouette_score(val_x, y_pred_random)
```

```python
log_likelihood_grid = best_model_grid.score(val_x)
log_likelihood_random = best_model_random.score(val_x)
```

**Resultados del GridSearchCV**

```python
print("\n Resultados Grid Search")
print(f"Tiempo de búsqueda: {grid_time:.2f} segundos")
print(f"Silhouette Score: {silhouette_grid:.4f}")
print(f"Log-Likelihood Score: {log_likelihood_grid:.4f}")
```

**Resultados del RandomizedSearchCV**

```python
print("\n Resultados Random Search")
print(f"Tiempo de búsqueda: {random_time:.2f} segundos")
print(f"Silhouette Score: {silhouette_random:.4f}")
print(f"Log-Likelihood Score: {log_likelihood_random:.4f}")
```

Antes de analizar los resultados toca hablar sobre las metricas. *Silhouette Score* mide que tan acertada es la agrupación dentro del cluster en comparación a sus vecinos, mientras mas cercano al 1 sea el valor mejor es el agrupamiento. Por su parte *Log-Likelihood* mide el ajuste del modelo a los datos.

Viendo los resultados, el modelo falla en ambas tareas. No agrupa los datos de manera optima ni se ajusta bien a los datos. En un escenario donde se busque obtener el mejor resultado, se aplicarian tecnicas de para mejorar el rendimiento del modelo como identificar componentes principales o mejorar la calidad de los datos eliminando outliers. En este caso, dado que solamente se busca comparar el rendimiento de los metodos de optimizacion y que ya se obtuvo un modelo que entrega buenos resultados con los algoritmos supervisado no voy a buscar mejorar el desempeño del modelo en este momento.

Lo que me parece un buen ejercicio es ver en detalle los resultados que mando el modelo, para tener una mejor idea de como funciona. Para ello, se agregan los resultado que da el modelo optimizado con *GridSearchCV* al dataset

```python
val_x["Cluster_GMM"] = best_model_grid.predict(val_x)
```

Podemos graficar como se reparten los cluesters respectoa variables correlacionadas 

```python
plt.scatter(val_x['E1'], val_x['pt1'], c=val_x["Cluster_GMM"], cmap="viridis", alpha=0.5)
plt.xlabel("E1 GeV")
plt.ylabel("pt1 GeV")
plt.title("Clusters obtenidos con GaussianMixture")
plt.colorbar(label="Cluster")
plt.show()
```

```python
plt.scatter(val_x['E2'], val_x['pt2'], c=val_x["Cluster_GMM"], cmap="viridis", alpha=0.5)
plt.xlabel("E2 GeV")
plt.ylabel("pt2 GeV")
plt.title("Clusters obtenidos con GaussianMixture")
plt.colorbar(label="Cluster")
plt.show()
```

En el primer gráfico no se distiguen grupos claramente separados. Si bien se puede interpretar que es debido a la correlación entre las variables, si se hace la comparación con el segundo gráfico donde la division es más evidente. Tal que como indicaba el resultado de rendimiento, el modelo no es del todo adecuado para este dataset. Ahora bien, otro analisis que pueden dar más información sobre el rendimiento es ver cuales son las probabilidades que asigna el modelo a cada punto

```python
probabilidades = best_model_grid.predict_proba(val_x.drop(columns = 'Cluster_GMM'))
probabilidades[:5]
```

Viendo los ordenes de magnitud, casi todas las posibilidaes son bajas, dando a enteder que el modelo tiene dificultades para determinar a que grupo pertenece cada punto. Lo cual es consistente con los resultados. 

Aun asi, con los modelos supervisados previos ya se obtuvo un buen resultado.
