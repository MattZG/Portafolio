### 2.1 Determine el tipo de variable de cada columna

```python
df.dtypes
```

### 2.2 Calcular la media, mediana y moda de la cantidad de medallas obtenidas

```python
print('La media de medallas obtenidas es: ', df['Medallas_Totales'].mean())
print('La mediana de medallas obtenidas es: ', df['Medallas_Totales'].median())
print('La moda de medallas obtenidas es: ', df['Medallas_Totales'].mode())
```

## 3. Análisis de Correlación

### 3.1 Calcular la correlacion de Pearson entre entrenamientos semanales y medallas totales.

```python
corr, p_value = pearsonr(df['Entrenamientos_Semanales'], df['Medallas_Totales'])
print('Correlacion: ', corr)
print('P-value: ', p_value)

if p_value < 0.05:
    print('La relacion entre las medallas obtenidas y los entrenamientos semanales es estadisticamente significativa')
else:
    print('La relacion entre las medallas obtenidas y los entrenamientos semanales no es estadisticamente significativa')
```

### 3.2 Crear un gráfico de dispersión entre peso y medallas totales

```python
plt.figure(figsize = (8, 6))
plt.scatter(df['Peso_kg'], df['Medallas_Totales'], color = 'black', marker = 'x')
plt.title('Relación entre el peso (Kg) y medallas obtenidas')
plt.xlabel('Peso en Kg')
plt.ylabel('Medallas')
plt.grid(True)
plt.show()
```

**No existe correlación entre estas variables, porque no hay una tendencia clara en el grafico.**

## 4. Regresión Lineal

### 4.1 Implementar un modelo de regresión lineal para predecir el n° de medallas obtenidas en función del n° de entrenamientos semanales

```python
x = df[['Entrenamientos_Semanales']]
y = df['Medallas_Totales']
model = LinearRegression()
model.fit(x, y)
```

### 4.2 Obtener los coeficientes de regresión e interpretar el resultado

```python
print('El valor del intercepto es: ' , model.intercept_)
print('El valor del coefciente es: ', model.coef_[0])
```

Hay una relación creciente entre el numero de entrenamientos semanales y la obtención de medallas. Por lo tanto, se puede inferir a que a mayor número de entrenamientos semanales es mayor la posibilidad de obtener una medalla.

### 4.3 Calcular R2

```python
y_pred = model.predict(x)
print('R2: ', r2_score(y, y_pred))
```

### 4.4 Use *regplot* para graficiar la regresión lineal

```python
plt.figure(figsize = (8, 6))
sns.regplot(data = df, x = 'Entrenamientos_Semanales', y = 'Medallas_Totales', color = 'red', marker = 'x', line_kws = dict(color = 'blue'))
plt.title('Relación lineal entre entrenamientos semanales y medallas obtenidas')
plt.xlabel('Número de entrenamientos semanales')
plt.ylabel('Número de medallas obtenidas')
plt.show()
```
