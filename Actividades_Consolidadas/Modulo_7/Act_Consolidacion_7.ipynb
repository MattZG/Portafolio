## 2. Aplicación de algoritmos de clusterización

Preparación de los datos para entrenar el modelo

```python
df_sin_pais = df.drop(columns = ['País'])
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_sin_pais), 
                        columns = df_sin_pais.columns, 
                        index = df.index)
```

Aplicando K-Means con k=3 inicial

```python
kmeans_3 = KMeans(n_clusters = 3, random_state = 42, n_init = 10)
clusters_kmeans_3 = kmeans_3.fit_predict(df_scaled)

# Añadir clusters al dataframe
df['cluster_kmeans_3'] = clusters_kmeans_3

print("Distribución de países por cluster (K-Means k=3):")
for i in range(3):
    countries_in_cluster = df[df['cluster_kmeans_3'] == i].index.tolist()
    print(f"Cluster {i}: {countries_in_cluster}")
```

Metodo del codo para determinar el k optimo

```python
k_range = range(2, 8)
inertias = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10)
    clusters = kmeans.fit_predict(df_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_scaled, clusters))

# Visualizar método del codo y coeficiente de silueta
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))

# Método del codo
ax1.plot(k_range, inertias, 'bo-')
ax1.set_xlabel('Número de clusters (k)')
ax1.set_ylabel('Inercia (WCSS)')
ax1.set_title('Método del Codo')
ax1.grid(True)

# Encontrar el codo automáticamente
kn = KneeLocator(k_range, inertias, curve = 'convex', direction = 'decreasing')
optimal_k_elbow = kn.elbow

# Coeficiente de silueta
ax2.plot(k_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Número de clusters (k)')
ax2.set_ylabel('Coeficiente de Silueta')
ax2.set_title('Análisis de Silueta')
ax2.grid(True)

optimal_k_silhouette = k_range[np.argmax(silhouette_scores)]

plt.tight_layout()
plt.show()

print(f"K óptimo según método del codo: {optimal_k_elbow}")
print(f"K óptimo según coeficiente de silueta: {optimal_k_silhouette}")
```

Aplicar K-Means con k óptimo

```python
optimal_k = optimal_k_silhouette  # Usar el k con mejor silueta
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters_kmeans_optimal = kmeans_optimal.fit_predict(df_scaled)
df['cluster_kmeans_optimal'] = clusters_kmeans_optimal

print(f"\nDistribución con k óptimo ({optimal_k}):")
for i in range(optimal_k):
    countries_in_cluster = df[df['cluster_kmeans_optimal'] == i].index.tolist()
    print(f"Cluster {i}: {countries_in_cluster}")
```

Clustering jerarquico

```python
linkage_matrix = linkage(df_scaled, method='ward')

plt.figure(figsize=(15, 10))
dendrogram(linkage_matrix, labels=df.index, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Países')
plt.ylabel('Distancia')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Aplicar clustering jerárquico con el número óptimo de clusters
hierarchical = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward')
clusters_hierarchical = hierarchical.fit_predict(df_scaled)
df['cluster_hierarchical'] = clusters_hierarchical

print("Distribución por clustering jerárquico:")
for i in range(optimal_k):
    countries_in_cluster = df[df['cluster_hierarchical'] == i].index.tolist()
    print(f"Cluster {i}: {countries_in_cluster}")

# Comparar K-Means vs Jerárquico
agreement = np.sum(clusters_kmeans_optimal == clusters_hierarchical) / len(clusters_kmeans_optimal)
print(f"\nAcuerdo entre K-Means y Clustering Jerárquico: {agreement:.2%}")
```

DBSCAN

```python
eps_values = [0.3, 0.5, 0.7, 1.0, 1.2]
min_samples_values = [3, 5, 8]

best_eps = None
best_min_samples = None
best_score = -1
best_n_clusters = 0

print("Probando diferentes parámetros para DBSCAN:")
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters_dbscan = dbscan.fit_predict(df_scaled)
        
        n_clusters = len(set(clusters_dbscan)) - (1 if -1 in clusters_dbscan else 0)
        n_noise = list(clusters_dbscan).count(-1)
        
        if n_clusters > 1:  # Solo calcular silueta si hay más de 1 cluster
            # Excluir puntos de ruido para el cálculo de silueta
            mask = clusters_dbscan != -1
            if mask.sum() > 1:
                score = silhouette_score(df_scaled[mask], clusters_dbscan[mask])
                print(f"eps={eps}, min_samples={min_samples}: {n_clusters} clusters, {n_noise} ruido, silueta={score:.3f}")
                
                if score > best_score:
                    best_score = score
                    best_eps = eps
                    best_min_samples = min_samples
                    best_n_clusters = n_clusters
            else:
                print(f"eps={eps}, min_samples={min_samples}: {n_clusters} clusters, {n_noise} ruido, silueta=N/A")
        else:
            print(f"eps={eps}, min_samples={min_samples}: {n_clusters} clusters, {n_noise} ruido")

# Aplicar DBSCAN con mejores parámetros
if best_eps is not None:
    dbscan_best = DBSCAN(eps=best_eps, min_samples=best_min_samples)
    clusters_dbscan_best = dbscan_best.fit_predict(df_scaled)
    df['cluster_dbscan'] = clusters_dbscan_best
    
    print(f"\nMejores parámetros DBSCAN: eps={best_eps}, min_samples={best_min_samples}")
    print(f"Número de clusters: {best_n_clusters}")
    print(f"Coeficiente de silueta: {best_score:.3f}")
    
    # Mostrar composición de clusters
    for cluster_id in sorted(set(clusters_dbscan_best)):
        if cluster_id == -1:
            countries_in_cluster = df[df['cluster_dbscan'] == cluster_id].index.tolist()
            print(f"Ruido: {countries_in_cluster}")
        else:
            countries_in_cluster = df[df['cluster_dbscan'] == cluster_id].index.tolist()
            print(f"Cluster {cluster_id}: {countries_in_cluster}")
else:
    print("No se encontraron parámetros adecuados para DBSCAN")
    df['cluster_dbscan'] = -1
```

## 3. Aplicación de reducción de dimensionalidad

PCA

```python
pca_full = PCA()
pca_full.fit(df_scaled)
explained_variance_ratio = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Encontrar número de componentes para 90% de varianza
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
print(f"Número de componentes principales para explicar 90% de la varianza: {n_components_90}")

# Visualizar varianza explicada
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.title('Varianza Explicada por Componente')

plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 'bo-')
plt.axhline(y=0.9, color='r', linestyle='--', label='90% varianza')
plt.axvline(x=n_components_90, color='r', linestyle='--', label=f'PC{n_components_90}')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Acumulada')
plt.title('Varianza Acumulada')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(df_scaled)

print(f"Varianza explicada por los primeros 2 componentes: {pca_2d.explained_variance_ratio_.sum():.3f}")

# Visualizar países en espacio PCA
plt.figure(figsize=(12, 10))
scatter = plt.scatter(pca_2d_result[:, 0], pca_2d_result[:, 1], 
                     c=df['cluster_kmeans_optimal'], cmap='tab10', s=100, alpha=0.7)
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.2%} varianza)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.2%} varianza)')
plt.title('Países en Espacio PCA (2D) - Coloreados por Cluster K-Means')
plt.colorbar(scatter)

# Añadir etiquetas de países
for i, country in enumerate(df.index):
    plt.annotate(country, (pca_2d_result[i, 0], pca_2d_result[i, 1]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8, alpha=0.7)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
cluster_columns = ['cluster_kmeans_3', 'cluster_kmeans_optimal', 'cluster_hierarchical', 'cluster_dbscan']
excluded_columns = ['País'] + cluster_columns

# Estas son tus características musicales reales
feature_names = [col for col in df.columns if col not in excluded_columns]

print("Características musicales identificadas:")
print(feature_names)
print(f"Número de características: {len(feature_names)}")

# Crear el DataFrame de componentes principales
components_df = pd.DataFrame(
    pca_2d.components_.T,  # Transponer para tener características en filas
    columns=['PC1', 'PC2'],
    index=feature_names
)

print("\nCaracterísticas más importantes en cada componente principal:")
print("PC1:")
print(components_df['PC1'].sort_values(key=abs, ascending=False))
print("\nPC2:")
print(components_df['PC2'].sort_values(key=abs, ascending=False))
```

Aplicando t-SNE

```python
perplexity_values = [1, 2, 3, 5]

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, perplexity in enumerate(perplexity_values):
    if i < len(axes):
        print(f"Aplicando t-SNE con perplexity={perplexity}...")
        tsne = TSNE(n_components = 2, perplexity = perplexity, random_state = 42, n_iter = 1000, learning_rate = 'auto', init='pca', method='exact')
        tsne_result = tsne.fit_transform(df_scaled)
        
        scatter = axes[i].scatter(tsne_result[:, 0], tsne_result[:, 1], 
                                 c=df['cluster_kmeans_optimal'], cmap='tab10', s=60, alpha=0.7)
        axes[i].set_title(f't-SNE (perplexity={perplexity})')
        axes[i].grid(True, alpha=0.3)
        
        # Añadir algunas etiquetas para los países más separados
        for j, country in enumerate(df.index[::4]):  # Mostrar cada 4to país para no saturar
            axes[i].annotate(country, (tsne_result[j*4, 0], tsne_result[j*4, 1]), 
                           xytext=(2, 2), textcoords='offset points', fontsize=6, alpha=0.7)

# Eliminar subplot vacío
if len(perplexity_values) < len(axes):
    fig.delaxes(axes[len(perplexity_values)])

plt.tight_layout()
plt.show()
```

## 4. Análisis de resultados y conclusiones

Comparación de algoritmos de clustering

```python
kmeans_silhouette = silhouette_score(df_scaled, df['cluster_kmeans_optimal'])
hierarchical_silhouette = silhouette_score(df_scaled, df['cluster_hierarchical'])

print(f"K-Means (k={optimal_k}):")
print(f"  - Coeficiente de silueta: {kmeans_silhouette:.3f}")
print(f"  - Número de clusters: {optimal_k}")
print(f"  - Ventajas: Rápido, eficiente, clusters esféricos")
print(f"  - Desventajas: Requiere especificar k, asume clusters esféricos")

print(f"\nClustering Jerárquico:")
print(f"  - Coeficiente de silueta: {hierarchical_silhouette:.3f}")
print(f"  - Número de clusters: {optimal_k}")
print(f"  - Ventajas: No requiere k previo, dendrograma interpretable")
print(f"  - Desventajas: Computacionalmente costoso, sensible a ruido")

if best_eps is not None:
    mask = df['cluster_dbscan'] != -1
    if mask.sum() > 1:
        dbscan_silhouette = silhouette_score(df_scaled[mask], df['cluster_dbscan'][mask])
        n_noise_final = (df['cluster_dbscan'] == -1).sum()
        print(f"\nDBSCAN (eps={best_eps}, min_samples={best_min_samples}):")
        print(f"  - Coeficiente de silueta: {dbscan_silhouette:.3f}")
        print(f"  - Número de clusters: {best_n_clusters}")
        print(f"  - Países clasificados como ruido: {n_noise_final}")
        print(f"  - Ventajas: Encuentra clusters de forma arbitraria, identifica outliers")
        print(f"  - Desventajas: Sensible a parámetros, puede no encontrar todos los clusters")

# Determinar el mejor algoritmo
best_algorithm = 'K-Means'
best_score = kmeans_silhouette

if hierarchical_silhouette > best_score:
    best_algorithm = 'Clustering Jerárquico'
    best_score = hierarchical_silhouette

if best_eps is not None and mask.sum() > 1:
    if dbscan_silhouette > best_score:
        best_algorithm = 'DBSCAN'
        best_score = dbscan_silhouette

print(f'\n MEJOR ALGORITMO: {best_algorithm} (silueta = {best_score:.3f})')
```
